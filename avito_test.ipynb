#### <span style="color:red;"> Задача 6 </span>
Предположим, у Шерлока Холмса есть 27 бюстов Наполеона. Вы знаете, что один из них фальшивый и отличается по весу от остальных. За какое минимальное количество взвешиваний на чашечных весах без гирь можно наверняка определить легче он или тяжелее?

**Ответ: 4 взвешивания.**
- Делим бюсты на 3 группы по 9 штук. Первое взвешивание: ставим на весы две группы, далее возможно два варианта: если весы не отклоняются, то фальшивый в третьей группе. Если отклонились, то он в одной из взвешенных групп и либо легче, либо тяжелее:
    1. Весы равны. Берём третью группу и делим 3x3, делаем второе взвешивание, снова получаем два варианта:
        - Весы равны. Остаётся три бюста. Делаем третье взвешивание, весы снова равны - фальшивый третий. На четвёртое взвешивание определяем легче он или тяжелее.
        - .....
    2. Весы отклонились. Значит фальшивый во взятых группах. Выбираем подозреваюмую группу, допустим она легче, делим 3х3, делаем второе взвешивание.
        - .....
        - Весы отклонились. Выбираем снова более лёгкую. Делаем третье взвешивание. Четвёртым взвешиванием либо находим более лёгкий фальшивый, либо понимаем, что фальшивый был в другой группе и он более тяжёлый.
- Есть 27 бюстов, каждый из которых может быть либо легче, либо тяжелее, тогда всего возможных сценариев `27 х 2 = 54`. Каждое взвешивание даёт 3 возможных исхода. За k взвешиваний можно различить `3**k` сценариев, следовательно должно выполняться `3**k >= 54`, тогда минимальное `k = 4`.

#### <span style="color:red;"> Задача 7 </span>

Предварительные тесты показали, что 30% из них очень хорошо готовы к экзамену, 50% достаточно хорошо и 20% кое-как. Для первой группы веротность получение отлично - 90%, для второй - 70%, для третьей - 30%. Случайный студент получает отлично. Какова вероятность, что он из первой группы?

**Ответ: 27/68**
- Классическая задача на теорему Байеса:
    - событие A1 - студент из первой группы, `P(A1) = 0.3`
    - `P(A2) = 0.5`
    - `P(A3) = 0.2`
    - событие B - случайный студент получил отлично
    - `P(B|A1) = 0.9`
    - `P(B|A2) = 0.7`
    - `P(B|A3) = 0.3`
    - Найти `P(A1|B)`
- Используем формулу полной вероятности: `P(B) = P(B∣A1​)*P(A1​) + P(B∣A2​)*P(A2​) + P(B∣A3​)P(A3​) = 0.9*0.3 + 0.7*0.5 + 0.3*0.2 = 0.68`
- Используем формулу Байеса: $ P(A_1 \mid B) = \frac{P(B \mid A_1) \cdot P(A_1)}{P(B)} $
- Получаем `P(A1|B) = (0.9 * 0.3) / 0.68 = 27/68`

#### <span style="color:red;"> Задача 16 </span>
Плотность вероятности (PDF(x)) - один из способов задания распределения непрерывной случайной величины.<br><br>
$ pdf_X(x) = \lim_{\Delta x \to 0} \frac{P(X \in [x, x + \Delta x])}{\Delta x} $<br><br>
Сколько утверждений верно в этом списке?
- PDF(x) достигает своего максимума в точке х = Ex
- PDF(x) = 1 для всех х
- если P(x=a) = 0, то PDF(а) = 0

**Ответ: 0**
- Для нормального распределения максимум плотности совпадает с математическим ожиданием, но для экспоненциального распределения максимум плотности находится в нуле, а `Ex = 1/λ`.
- Единице должна быть равна суммарная вероятность всех значений, то есть интеграл функции, а её значение не обязано быть равно 1.
- Для непрерывной случайной величины вероятность того, что она примет конкретное значение `a`, всегда равна нулю. Но `PDF(a)` может быть любым неотрицательным значением. Например, в нормальном распределении `PDF(a)` максимальна в среднем значении, но `P(a)` все равно равно нулю.

#### <span style="color:red;"> Задача 18 </span>
Есть случайная величина X, которая распределена следующим образом: X может принимать любые неотрицательные значения и `P(X=k) = 1/(2^(k+2))` для любого k>0. Найти `P(X=0)`.

**Ответ: 3/4**
- Сумма всех вероятностей для всех возможных значений X должна равняться 1, так как это основное свойство распределения вероятностей.<br><br>
$ P(X=0) + \sum_{k=1}^{\infty} P(X=k) = 1 $ <br><br>
- Нам уже даны вероятности для k>0, подставим в формулу.<br><br>
$ P(X=0) + \sum_{k=1}^{\infty} \frac{1}{2^{k+2}} = 1 $<br><br>
- Вынесем `1/2**2 = 1/4` за знак суммы.<br><br>
$ \sum_{k=1}^{\infty} \frac{1}{2^{k+2}} = \frac{1}{4} \sum_{k=1}^{\infty} \left(\frac{1}{2}\right)^k $<br><br>
- Внутри осталась геометрическая прогрессия с первым членом 1/2​ и знаменателем 1/2.​<br><br>
$ \sum_{k=1}^{\infty} \left(\frac{1}{2}\right)^k = \frac{\frac{1}{2}}{1 - \frac{1}{2}} = 1 $​<br><br>
- Итоговая сумма.​<br><br>
$ \frac{1}{4} \cdot 1 = \frac{1}{4} $
- Находим \( P(X=0) \).​<br><br>
$ P(X=0) = 1 - \frac{1}{4} = \frac{3}{4} $​<br><br>

#### <span style="color:red;"> Задача 19 </span>
В коробке 60 белых и 40 черных шаров. Берётся 20 случайных. Найти ожидаемое значение количества черных среди них.

**Ответ: 8**
- Для гипергеометрического распределения (выборка без возвращения) математическое ожидание количества успехов (в данном случае — черных шаров) вычисляется по формуле: `Ex = n * (k/N)`, где n - размер выборки, k - общее количество успехов, N - размер совокупности. Тогда `Ex = 20 * (40/100) = 8`.

#### <span style="color:red;"> Задача 20 </span>
Для какого метода проверки гипотез необходимо знание о генеральной совокупности?
- ни для одного
- z-test
- тест Манна-Уитни
- t-test

**Ответ: z-test**
- Тест Манна-Уитни — это непараметрический тест, который используется для сравнения двух независимых выборок. Он основан на рангах, не предполагает нормального распределения данных и не требует знания параметров генеральной совокупности. Значит, для него знание о совокупности не нужно.
- t-test — этот тест используется, когда дисперсия генеральной совокупности неизвестна. Вместо неё используется выборочная дисперсия. Таким образом, t-test не требует знания о параметрах генеральной совокупности.
- z-test — этот тест используется, когда известны параметры генеральной совокупности: среднее и стандартное отклонение. Он применяется для проверки гипотез о среднем значении, особенно при больших объемах выборки. Если информация о генеральной совокупности отсутствует, z-test применять нельзя.

#### <span style="color:red;"> Задача 21 </span>
На этапе проектирования A/B теста при baseline конверсии = 50% и MDE = 1% (minimum detectable effect) рассчитали необходимую выборку для эксперимента. Во сколько раз нужно увеличить выборку, чтобы получить MDE = 0.1% ? Выберите ближайшее к правильному ответу.
- 10
- 100
- 3
- не нужно увеличивать, MDE от размера выборки не зависит

**Ответ: 100**
- Baseline конверсия - это исходный процент пользователей, которые выполняют целевое действие в текущих условиях (до внедрения изменений).
- MDE - это минимальное изменение конверсии, которое вы хотите обнаружить с помощью A/B-теста:
    - если baseline = 50%, а MDE = 1%, это означает, что вы хотите зафиксировать изменение конверсии от 50% до 51% (или до 49%).
- Размер выборки для A/B-теста обратно пропорционален квадрату MDE, это означает, что если мы уменьшаем MDE в k раз, то размер выборки нужно увеличить в k**2 раз.
    - если MDE уменьшается в 10 раз, то требуемый размер выборки увеличивается в 100 раз.

#### <span style="color:red;"> Задача 22 </span>
Подходит ли t-test для оценки результатов эксперимента, если целевая метрика имеет распределение Бернулли?
- Да
- Нет
- Да, но с обязательной предварительной линеаризацией

**Ответ: да**
- Распределение Бернулли — это дискретное распределение с двумя исходами: успех (1) и неудача (0).
- Обычно t-test используется для сравнения средних двух групп, предполагая, что данные распределены нормально или выборка достаточно большая для применения центральной предельной теоремы. Распределение Бернулли не является нормальным, особенно при малых вероятностях успеха (например, конверсия 1%). В таких случаях распределение будет сильно скошено.
Однако, если выборка достаточно большая, центральная предельная теорема говорит, что выборочное среднее будет стремиться к нормальному распределению. То есть, даже если исходные данные бернуллиевские, при больших n среднее приближается к нормальному распределению, и t-test может быть применим.
- Линеаризация — это метод преобразования целевой метрики так, чтобы уменьшить её дисперсию за счёт учёта влияющих факторов. Это позволяет сделать статистические тесты более чувствительными и точными. Но сама по себе она не обязательна для применения t-теста на данных с распределением Бернулли.

#### <span style="color:red;"> Задача 39 </span>
У нас есть две коробки с конфетами: в первой 20 конфет, из которых 3 ванильные, а во второй 40 конфет, из которых 2 ванильные. Мы выбираем одну коробку случайным образом, достаем конфету, и она оказывается ванильной. Нужно найти вероятность того, что мы выбрали вторую коробку.

**Ответ: 0.25**
- P(B1) — вероятность выбрать первую коробку.
- P(B2) — вероятность выбрать вторую коробку.
- P(V|B1) — вероятность достать ванильную конфету из первой коробки.
- P(V|B2) — вероятность достать ванильную конфету из второй коробки.
- Поскольку выбор коробки равновероятен, P(B1) = P(B2) = 0.5.
- В первой коробке 3 ванильные конфеты из 20, значит P(V|B1) = 3/20 = 0.15.
- Во второй коробке 2 ванильные конфеты из 40, значит P(V|B2) = 2/40 = 0.05.
- `P(B2|V) = [P(V|B2) * P(B2)] / [P(V|B1) * P(B1) + P(V|B2) * P(B2)] = (0.05 * 0.5) / [(0.15 * 0.5) + (0.05 * 0.5)] = 0.25`.

#### <span style="color:red;"> Задача 43 </span>
Дискретная случайная величина может принимать только два значения a и b: b > a. P(X=a) = 0.2, EX = 2.6, VarX = 0.64. Найти a.

**Ответ: 1**
- `E[X] = a * P(X=a) + b * P(X=b)`
- `VarX = E[X²] - (E[X])²`
- `E[X²] = a² * P(X=a) + b² * P(X=b)`
- Так как случайная величина принимает только два значения, и сумма вероятностей должна быть равна 1, то `P(X=b) = 1 - P(X=a) = 1 - 0.2 = 0.8`
- Теперь подставим известные значения в формулу математического ожидания: `2.6 = a * 0.2 + b * 0.8` Это уравнение связывает a и b.
- Теперь перейдем к дисперсии. Сначала найдем `E[X²] = a² * 0.2 + b² * 0.8` <=> `0.64 = (a² * 0.2 + b² * 0.8) - (2.6)²` <=> `0.2a² + 0.8b² = 7.4`
- Теперь у нас есть система из двух уравнений:<br>
  `0.2a + 0.8b = 2.6`<br>
  `0.2a² + 0.8b² = 7.4`
- Для решения этой системы можно выразить одну переменную через другую из первого уравнения и подставить во второе. Например, выразим b из первого уравнения:<br>
  `b = (2.6 - 0.2a) / 0.8 = 3.25 - 0.25a`
- Теперь подставим это выражение для b во второе уравнение: `0.2a² + 0.8*(3.25 - 0.25a)² = 7.4` <=> `0.2a² + 0.8*(10.5625 - 1.625a + 0.0625a²) = 7.4` <=><br>
  `0.25a² - 1.3a + 1.05 = 0` <=> `a² - 5.2a + 4.2 = 0` => `a1 = (5.2 + 3.2)/2 = 8.4/2 = 4.2` `a2 = (5.2 - 3.2)/2 = 2/2 = 1`
- Теперь проверим оба решения. По условию, b > a. Найдем b для каждого a: `b1 = 3.25 - 0.25*4.2 = 3.25 - 1.05 = 2.2` `b2 = 3.25 - 0.25*1 = 3.25 - 0.25 = 3`
- `b = 3 > a = 1`
